```python
df1=pd.read_csv('/content/Actuators.csv')
df2=pd.read_csv('/content/Physical.csv')
df3=pd.read_csv('/content/Pilot.csv')
```

```python
df1.dropna(how='all',inplace=True)
df2.dropna(how='all',inplace=True)
df3.dropna(how='all',inplace=True)
```

```python
x1=df1.drop(['label'],axis=1)
y1=df1.label.values
```

```python
x2=df2.drop(['label'],axis=1)
y2=df2.label.values
```

```python
x3=df3.drop(['label'],axis=1)
y3=df3.label.values
```

### **Training and testing**

```python
from sklearn.model_selection import train_test_split

x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,test_size=0.25)
x2_train,x2_test,y2_train,y2_test=train_test_split(x2,y2,test_size=0.25)
x3_train,x3_test,y3_train,y3_test=train_test_split(x3,y3,test_size=0.25)
```

```python
len(x1_train)
len(x2_train)
len(x2_train)
```

```python
len(x1_test)
len(x2_test)
len(x2_test)
```

### **Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
x1_train=sc.fit_transform(x1_train)
x1_test=sc.transform(x1_test)

x2_train=sc.fit_transform(x2_train)
x2_test=sc.transform(x2_test)

x3_train=sc.fit_transform(x3_train)
x3_test=sc.transform(x3_test)
```

### **Model Building**

```python
from sklearn.neighbors import KNeighborsClassifier

model_knn=KNeighborsClassifier()
model_knn.fit(x1_train,y1_train)
```

```python
y1_pred_knn=model_knn.predict(x1_test)
```

```python
y1_pred_knn
```

```python
y1_test
```

## **Classification Metrics**

```python
from sklearn.metrics import accuracy_score

print(round(accuracy_score(y1_test,y1_pred_knn)*100,2))
```
